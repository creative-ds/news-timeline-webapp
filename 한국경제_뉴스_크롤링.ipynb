```python
from bs4 import BeautifulSoup
from datetime import datetime
import requests
import pandas as pd
import re

from datetime import datetime, timedelta
from tqdm.notebook import tqdm
```

```python
def date_to_str(date):
    y = str(date.year)
    m = str(date.month).zfill(2)
    d = str(date.day).zfill(2)

    str_date = y + '.' + m + '.' + d + '.'
    return str_date
```

```python
# 크롤링할 뉴스 기사 기간 설정
s_date = datetime(2021, 2, 26)
e_date = datetime(2021, 8, 31)

date = s_date

# 크롤링 결과 저장하기 위한 리스트 선언 
title_text = []
link_text = []
date_text = []
body_text = []
result = {}

while date <= e_date:
    str_date = date_to_str(date)
    print(str_date)

    tmp_link = [] 
    page=1
    while True:
        url = "https://www.hankyung.com/society/1001?date=" + str_date + "&page=" + str(page)

        response = requests.get(url)
        html = response.text

        # 뷰티풀소프의 인자값 지정
        soup = BeautifulSoup(html, 'html.parser')

        # <a>태그에서 제목과 링크주소 추출
        atags = soup.select('.tit a')
        if len(atags) == 0: break

        for atag in atags:
            title_text.append(atag.text)  # 제목
            tmp_link.append(atag['href'])  # 링크
        
        page += 1

    date = date + timedelta(1)
    link_text += tmp_link

    for link in tqdm(tmp_link):
        response = requests.get(link, headers={'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.9; rv:50.0) Gecko/20100101 Firefox/50.0'})
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')
        if len(soup.find_all('div', id ='articletxt')) == 0:
            body_text.append('웹 페이지를 찾을 수 없습니다.')
            date_text.append(str_date[:-1])
            continue

        body = soup.find_all('div', id ='articletxt')[0].text
        body_text.append(body)
        date_text.append(str_date[:-1])

# 모든 리스트 딕셔너리형태로 저장
result = {"date": date_text, "title": title_text, "body": body_text, "URL": link_text}

df = pd.DataFrame(result)  # df로 변환
```
